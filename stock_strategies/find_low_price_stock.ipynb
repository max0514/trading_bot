```python

# Assign price_df to df for further processing
df = price_df

# Compute the 15th percentile (quantile) of each row (axis=1) in the DataFrame df
# This operation will return a Series where each value corresponds to the 15th percentile of each row in df
q1 = df.quantile(0.15, axis=1)

# Create a new DataFrame from the values of the Series q1
# [q1.values] * len(df.columns) creates a list of arrays where q1.values is repeated
# The repetition count is equal to the number of columns in df
# The transpose() method is then used to set index to columns and set column to index
# where rows correspond to original q1 values and columns correspond to those in df
q1_df = pd.DataFrame([q1.values] * len(df.columns)).transpose()

# Set the column names of q1_df to match those of df
# This is important to ensure that the columns in both DataFrames are aligned for any future operations
q1_df.columns = df.columns

# Set the index of q1_df to match the index of df
# This aligns the rows of q1_df with those of df, maintaining the consistency of data across both DataFrames
q1_df.index = df.index


#if the column and index is not align, it will have error: index must be indentically labeled
#pick price < 15% of overall stock price
condition_1 = df<q1_df
condition_1

condition_2 = df <= 15

condition_3 = (1 -(df.rolling(60).min() / df.rolling(60).max())) < 0.3




#conditions = (condition_1 & condition_2 & condition_3)
conditions = (condition_1 & condition_2)
position = df[conditions]
position
report =sim(position=position, resample='M', trade_at_price='open',position_limit=1/5, fee_ratio=1.425/1000*0.3, stop_loss=0.3655)
report.display_mae_mfe_analysis()
```
